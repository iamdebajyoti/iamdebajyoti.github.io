```python
##### STEP 0 #####
##### AD REPORT COMPILATION #####
os.chdir(r"C:\DEB\DXC_Automation\UCMS_29864\_input_")    
df_Compiled_AD_files = pd.DataFrame()


file_name = ""
for files_ in (os.listdir(r"C:\DEB\DXC_Automation\UCMS_29864\_input_")):
    df_Temp = pd.DataFrame()
#     print(files_)
    Temp = pd.read_csv(files_, iterator=True, low_memory=False)
    df_Temp = pd.concat(Temp, ignore_index=True)
    
    df_Compiled_AD_files = pd.concat([df_Compiled_AD_files, df_Temp], ignore_index=True)
    print("The file {} has {} rows and the compiled file has {} rows.".format(files_,df_Temp.shape[0], df_Compiled_AD_files.shape[0]))
print("Verify: The compiled AD files contains {} rows.".format(df_Compiled_AD_files.shape[0]))
df_Compiled_AD_files.info()
sorted(df_Compiled_AD_files)
```
